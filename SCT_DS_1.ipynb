```python
selected_countries = ['India', 'China', 'United States', 'Indonesia', 'Brazil', 'Pakistan', 'Nigeria', 'Bangladesh', 'Russia', 'Mexico']
data = df[df['Country Name'].isin(selected_countries)][['Country Name', '2022']]

plt.figure(figsize=(10, 6))
plt.bar(data['Country Name'], data['2022'], color='skyblue')
plt.title('Population of Selected Countries in 2022')
plt.xlabel('Country')
plt.ylabel('Population')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```
